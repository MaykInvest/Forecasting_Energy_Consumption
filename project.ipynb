````python
# =====================================================================
# PHASE 1: IN-MEMORY DISTRIBUTED DATA GENERATION (WINDOWS COMPATIBLE)
# =====================================================================

import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, DoubleType, StringType

print("⏳ Generating production-realistic dataset directly inside Spark RAM...")

# 1. Create a baseline sequence of timestamps for 2 full years (17,520 hours)
start_timestamp = "2024-01-01 00:00:00"

df_base = spark.range(0, 17520).withColumn(
    "Timestamp", 
    F.to_timestamp(F.lit(start_timestamp)) + (F.col("id") * F.expr("INTERVAL 1 HOUR"))
).drop("id")

# 2. Add structural calendar features
df_features = df_base \
    .withColumn("Hour", F.hour(F.col("Timestamp"))) \
    .withColumn("Month", F.month(F.col("Timestamp"))) \
    .withColumn("Is_Weekend", F.when(F.dayofweek(F.col("Timestamp")).isin(1, 7), 1).otherwise(0))

# 3. Mathematically model realistic Weather (Temperature) and Energy Loads
df_modeled = df_features \
    .withColumn("Base_Temp", 15.0 + 12.0 * F.sin(2.0 * 3.14159 * (F.col("Month") - 4) / 12.0)) \
    .withColumn("Daily_Noise", (F.rand(seed=42) * 6.0) - 3.0) \
    .withColumn("Temperature_C", F.round(F.col("Base_Temp") + F.col("Daily_Noise") + (3.0 * F.sin(2.0 * 3.14159 * (F.col("Hour") - 6) / 24.0)), 1)) \
    .withColumn("Temp_Effect", 2.5 * F.abs(F.col("Temperature_C") - 18.0)) \
    .withColumn("Hour_Effect", 15.0 * F.sin(2.0 * 3.14159 * (F.col("Hour") - 6) / 24.0)) \
    .withColumn("Weekend_Reduction", F.col("Is_Weekend") * -12.0)

df_clean = df_modeled.withColumn(
    "Clean_Consumption",
    F.round(50.0 + F.col("Temp_Effect") + F.col("Hour_Effect") + F.col("Weekend_Reduction") + (F.rand(seed=24) * 8.0 - 4.0), 2)
)

# 4. Inject real-world missing values and anomalies (The Mistakes)
df_dirty = df_clean \
    .withColumn(
        "Temperature_C", 
        F.when(F.rand(seed=1) < 0.05, F.lit(None)).otherwise(F.col("Temperature_C"))
    ) \
    .withColumn(
        "Energy_Consumption_kWh",
        F.when(F.rand(seed=2) < 0.03, F.lit(None))          # 3% Missing values
         .when(F.rand(seed=3) < 0.005, F.lit(-999.0))       # Anomaly: Negative sensor fault
         .when(F.rand(seed=4) < 0.005, F.lit(9999.0))       # Anomaly: Voltage surge spike
         .otherwise(F.col("Clean_Consumption"))
    ) \
    .withColumn(
        "Is_Weekend_Dirty",
        F.when(F.col("Is_Weekend") == 1, F.lit("YES"))
         .otherwise(F.when(F.rand(seed=5) < 0.02, F.lit("UNKNOWN")).otherwise(F.lit("NO")))
    )

# Select final columns and cache the DataFrame in memory for optimal speed
df = df_dirty.select("Timestamp", "Temperature_C", "Is_Weekend_Dirty", "Energy_Consumption_kWh").cache()

# Trigger an action to verify the data is successfully generated
total_rows = df.count()
print(f"\n✅ [SUCCESS] In-Memory Spark DataFrame Ready!")
print(f"📊 Total Rows Generated: {total_rows}")
print("-" * 50)

# Generate by Inteligence Artificial - ChatGPT
````

### **📊 Core Dataset Feature Specifications**

Before feeding our historical records into distributed machine learning architectures, we must define our system variables. Below is the structural data dictionary mapping our tracking metrics:

* **`Timestamp` [Data Type: Timestamp]**
    * **Definition:** The precise calendar date and localized hour index ($24$-hour format) of the data collection cycle.
    * **Strategic Role:** Serves as our primary time-series timeline. It allows the Spark engine to extract temporal cyclical features like **Hour of the Day** and **Month of the Year**.

* **`Temperature_C` [Data Type: Double / Precision Decimal]**
    * **Definition:** The outdoor ambient weather temperature recorded during that hour, measured in degrees Celsius.
    * **Strategic Role:** This is our primary continuous **Numerical Feature (Input)**. Weather is the heaviest external variable driving thermodynamic grid demand (heating and cooling loads).

* **`Is_Weekend_Dirty` [Data Type: String / Categorical Text]**
    * **Definition:** A text flag indicator capturing whether the record occurred during standard weekend operation hours (`"YES"` for Saturday/Sunday, `"NO"` for weekdays).
    * **Strategic Role:** Our baseline **Categorical Feature (Input)**. It maps macro-level human behavior shifts, tracking when commercial/industrial offices shut down and residential power patterns take over.

* **`Energy_Consumption_kWh` [Data Type: Double / Target Variable]**
    * **Definition:** The total electrical power load utilized across the monitored grid sector during that single hour, measured in Kilowatt-hours (kWh).
    * **Strategic Role:** The absolute **Target Variable (Output Label)**. This is the exact variable our Linear Regression and Random Forest models are being trained to forecast.

````python
print("⏳ Executing data purification...")

# ---------------------------------------------------------------------
# DROP THE OUTLIERS
# ---------------------------------------------------------------------
# Keep only realistic hourly energy values, filtering out the sensor faults
df_filtered = dfc.filter(
    (F.col("Energy_Consumption_kWh") >= 0) & 
    (F.col("Energy_Consumption_kWh") <= 500))

# ---------------------------------------------------------------------
# PROFESSIONAL TEMPERATURE FORWARD-FILL
# ---------------------------------------------------------------------
# Create a timeline window sorted sequentially by our Timestamp
timeline_window = Window.orderBy("Timestamp")

# Look backward to find the 'last' known valid temperature and copy it forward
df_temp_fixed = df_filtered.withColumn(
    "Temperature_C",
    F.coalesce(
        F.col("Temperature_C"), 
        F.last(F.col("Temperature_C"), ignorenulls=True).over(timeline_window)
    )
)

# ---------------------------------------------------------------------
# FIX CATEGORICAL TEXT POLLUTION
# ---------------------------------------------------------------------
# Swap out the corrupt "UNKNOWN" rows with standard "NO"
df_cleaned = df_temp_fixed.withColumn(
    "Is_Weekend_Dirty",
    F.when(F.col("Is_Weekend_Dirty") == "UNKNOWN", "NO").otherwise(F.col("Is_Weekend_Dirty"))
).cache() # Cache keeps this clean dataframe stored safely in memory

print("✅ [SUCCESS] Dataset successfully cleaned using enterprise time-series rules!")
print("-" * 60)

# Display a quick structural summary to confirm it's pristine
df_cleaned.summary().show()
````

##### 🔍📊 Exploratory Data Analysis (EDA)

**Before passing our big data arrays into complex machine learning pipelines, we must first decipher the hidden signatures within our data.**

**Exploratory Data Analysis (EDA) represents the foundational detective work of data science.** It is the critical phase where we strip away the noise of raw metrics to isolate the **underlying patterns, seasonal cycles, and structural behaviors** that govern our energy grid. In an enterprise deployment, training models blindly without running diagnostic analysis is a recipe for catastrophic prediction failure.

By leveraging **Apache PySpark’s distributed architecture**, we can aggregate massive time-series tracking windows across thousands of independent cluster partitions in seconds. In this phase of the project, we design and build a high-resolution visual dashboard focusing explicitly on **two critical operational vectors**:

* **1. The Temporal Footprint (Time Dynamics):** Mapping how electricity demand scales organically throughout a **24-hour human routine**, directly contrasting heavy afternoon industrial consumption against low, early-morning baselines.
* **2. The Thermodynamic Response (Weather Dynamics):** Isolating the highly sensitive, non-linear relationship between ambient outdoor temperatures and grid load. This explicitly maps the severe **"U-shape" weather response curve**—proving that extreme cold and extreme heat both trigger massive surges in heating and cooling demand.

Through these analytical lenses, we transform raw rows of numbers into **clear, actionable business intelligence**. This visual evidence establishes the exact mathematical justification for choosing **advanced non-linear machine learning architectures** like Random Forests over simple linear formulas.

````python
print("⏳ Computing distributed EDA summary tables...")

# ---------------------------------------------------------------------
# AGGREGATION 1: Hourly Trends (Energy Consumption by Hour of Day)
# ---------------------------------------------------------------------
# Extract the hour from our Timestamp, group by it, and calculate the average
df_hourly_trend = df_cleaned \
    .withColumn("Hour", F.hour(F.col("Timestamp"))) \
    .groupBy("Hour") \
    .agg(F.avg("Energy_Consumption_kWh").alias("Avg_Energy_kWh")) \
    .sort("Hour") \
    .toPandas() # Safely convert this tiny 24-row summary table to Pandas

# ---------------------------------------------------------------------
# AGGREGATION 2: Weather Impact (Energy Consumption by Temperature)
# ---------------------------------------------------------------------
# Round temperatures to the nearest whole degree to group similar weather profiles together
df_weather_trend = df_cleaned \
    .withColumn("Temp_Rounded", F.round(F.col("Temperature_C"), 0)) \
    .groupBy("Temp_Rounded") \
    .agg(F.avg("Energy_Consumption_kWh").alias("Avg_Energy_kWh")) \
    .sort("Temp_Rounded") \
    .toPandas() # Safely convert this small summary table to Pandas

print("✅ [SUCCESS] EDA Aggregations completed and cached locally for plotting!")
````

##### 🛠️⚙️ Feature Engineering Pipeline

Raw data profiles cannot be processed directly by core Machine Learning algorithms, as mathematical engines can only ingest purely numerical structures. Feature Engineering is the intentional translation layer where we transform our cleaned attributes into optimized mathematical arrays.

By utilizing Apache PySpark’s distributed Machine Learning library (`pyspark.ml`), we build an automated, cluster-optimized transformation pipeline. This process structurally encodes our categorical text fields into binary tokens ($1.0$ or $0.0$) and consolidates all independent variables into a single, unified feature vector space. Finally, we execute a deterministic $80/20$ data split, isolating pristine training arrays to educate our models and independent validation sets to accurately gauge our future predictive success.

````python
# =====================================================================
# PHASE 4: FEATURE ENGINEERING PIPELINE FOR MACHINE LEARNING
# =====================================================================

print("⏳ Initializing distributed feature engineering pipeline...")

# Step 1: Convert Categorical Text ("YES"/"NO") into Numeric Tokens (1.0 / 0.0)
indexer = StringIndexer(inputCol="Is_Weekend_Dirty", outputCol="Is_Weekend_Numeric")

# Step 2: Combine all input features into a single consolidated Vector column
# Our inputs for the model will be the Temperature and the Weekend indicator
feature_inputs = ["Temperature_C", "Is_Weekend_Numeric"]
assembler = VectorAssembler(inputCols=feature_inputs, outputCol="features")

# Step 3: Link the indexer and assembler together into an automated Pipeline
ml_pipeline = Pipeline(stages=[indexer, assembler])

# 🛠️ FIXED: Changed df_cleaned to df_cleaned to match our previous step
df_ml_ready = ml_pipeline.fit(df_cleaned).transform(df_cleaned)

# Step 4: Split the data into Training (80%) and Testing (20%) sets for verification
train_df, test_df = df_ml_ready.randomSplit([0.8, 0.2], seed=42)

print("✅ [SUCCESS] Feature Engineering Pipeline Completed Successfully!")
print(f"📉 Total Training Records: {train_df.count()}")
print(f"📈 Total Testing Records: {test_df.count()}")
print("-" * 60)

# Display our newly engineered 'features' vector column alongside our target
df_ml_ready.select("features", "Energy_Consumption_kWh").show(5, truncate=False)
````

````python
# =====================================================================
# PHASE 5: DISTRIBUTED MACHINE LEARNING MODEL TRAINING & EVALUATION
# =====================================================================

print("⏳ Spawning Machine Learning algorithms on the Spark Cluster...")

# ---------------------------------------------------------------------
# STEP 1: INITIALIZE THE MODELS
# ---------------------------------------------------------------------
# Set up our Linear Regression baseline
lr = LinearRegression(
    featuresCol="features", 
    labelCol="Energy_Consumption_kWh"
)

# Set up our Random Forest champion (using 50 parallel decision trees)
rf = RandomForestRegressor(
    featuresCol="features", 
    labelCol="Energy_Consumption_kWh", 
    numTrees=50, 
    seed=42
)

# ---------------------------------------------------------------------
# STEP 2: TRAIN THE MODELS ON THE TRAINING DATASET (80%)
# ---------------------------------------------------------------------
print("🏋️‍♂️ Training Linear Regression Model...")
lr_model = lr.fit(train_df)

print("🌲 Training Distributed Random Forest Model (This may take a brief moment)...")
rf_model = rf.fit(train_df)

# ---------------------------------------------------------------------
# STEP 3: GENERATE PREDICTIONS ON THE TEST DATASET (20%)
# ---------------------------------------------------------------------
print("🔮 Generating predictions on unseen test arrays...")
lr_predictions = lr_model.transform(test_df)
rf_predictions = rf_model.transform(test_df)

# ---------------------------------------------------------------------
# STEP 4: EVALUATE MODEL ERROR PERFORMANCE (RMSE)
# ---------------------------------------------------------------------
# Initialize our evaluator targeting our true label column
evaluator = RegressionEvaluator(
    labelCol="Energy_Consumption_kWh", 
    predictionCol="prediction", 
    metricName="rmse"
)

# Calculate Root Mean Squared Error for both models
lr_rmse = evaluator.evaluate(lr_predictions)
rf_rmse = evaluator.evaluate(rf_predictions)

# ---------------------------------------------------------------------
# STEP 5: PRINT THE FINAL PORTFOLIO PERFORMANCE LEADERBOARD
# ---------------------------------------------------------------------
print("\n" + "="*50)
print("🏆 MACHINE LEARNING MODEL LEADERBOARD 🏆")
print("="*50)
print(f"📉 Linear Regression Baseline RMSE : {lr_rmse:.2f} kWh")
print(f"🌲 Random Forest Champion    RMSE : {rf_rmse:.2f} kWh")
print("-" * 50)

# Calculate how much better the Random Forest performed
improvement = ((lr_rmse - rf_rmse) / lr_rmse) * 100
print(f"🚀 Random Forest reduced error rates by: {improvement:.1f}%")
print("="*50)

# Show a quick side-by-side preview of the actual values vs predictions
print("\n📊 Random Forest Predictions Quick-Look:")
rf_predictions.select("Temperature_C", "Energy_Consumption_kWh", "prediction").show(5)
````
