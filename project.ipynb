````python
# =====================================================================
# PHASE 1: IN-MEMORY DISTRIBUTED DATA GENERATION (WINDOWS COMPATIBLE)
# =====================================================================

import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, DoubleType, StringType

print("⏳ Generating production-realistic dataset directly inside Spark RAM...")

# 1. Create a baseline sequence of timestamps for 2 full years (17,520 hours)
start_timestamp = "2024-01-01 00:00:00"

df_base = spark.range(0, 17520).withColumn(
    "Timestamp", 
    F.to_timestamp(F.lit(start_timestamp)) + (F.col("id") * F.expr("INTERVAL 1 HOUR"))
).drop("id")

# 2. Add structural calendar features
df_features = df_base \
    .withColumn("Hour", F.hour(F.col("Timestamp"))) \
    .withColumn("Month", F.month(F.col("Timestamp"))) \
    .withColumn("Is_Weekend", F.when(F.dayofweek(F.col("Timestamp")).isin(1, 7), 1).otherwise(0))

# 3. Mathematically model realistic Weather (Temperature) and Energy Loads
df_modeled = df_features \
    .withColumn("Base_Temp", 15.0 + 12.0 * F.sin(2.0 * 3.14159 * (F.col("Month") - 4) / 12.0)) \
    .withColumn("Daily_Noise", (F.rand(seed=42) * 6.0) - 3.0) \
    .withColumn("Temperature_C", F.round(F.col("Base_Temp") + F.col("Daily_Noise") + (3.0 * F.sin(2.0 * 3.14159 * (F.col("Hour") - 6) / 24.0)), 1)) \
    .withColumn("Temp_Effect", 2.5 * F.abs(F.col("Temperature_C") - 18.0)) \
    .withColumn("Hour_Effect", 15.0 * F.sin(2.0 * 3.14159 * (F.col("Hour") - 6) / 24.0)) \
    .withColumn("Weekend_Reduction", F.col("Is_Weekend") * -12.0)

df_clean = df_modeled.withColumn(
    "Clean_Consumption",
    F.round(50.0 + F.col("Temp_Effect") + F.col("Hour_Effect") + F.col("Weekend_Reduction") + (F.rand(seed=24) * 8.0 - 4.0), 2)
)

# 4. Inject real-world missing values and anomalies (The Mistakes)
df_dirty = df_clean \
    .withColumn(
        "Temperature_C", 
        F.when(F.rand(seed=1) < 0.05, F.lit(None)).otherwise(F.col("Temperature_C"))
    ) \
    .withColumn(
        "Energy_Consumption_kWh",
        F.when(F.rand(seed=2) < 0.03, F.lit(None))          # 3% Missing values
         .when(F.rand(seed=3) < 0.005, F.lit(-999.0))       # Anomaly: Negative sensor fault
         .when(F.rand(seed=4) < 0.005, F.lit(9999.0))       # Anomaly: Voltage surge spike
         .otherwise(F.col("Clean_Consumption"))
    ) \
    .withColumn(
        "Is_Weekend_Dirty",
        F.when(F.col("Is_Weekend") == 1, F.lit("YES"))
         .otherwise(F.when(F.rand(seed=5) < 0.02, F.lit("UNKNOWN")).otherwise(F.lit("NO")))
    )

# Select final columns and cache the DataFrame in memory for optimal speed
df = df_dirty.select("Timestamp", "Temperature_C", "Is_Weekend_Dirty", "Energy_Consumption_kWh").cache()

# Trigger an action to verify the data is successfully generated
total_rows = df.count()
print(f"\n✅ [SUCCESS] In-Memory Spark DataFrame Ready!")
print(f"📊 Total Rows Generated: {total_rows}")
print("-" * 50)

# Generate by Inteligence Artificial - ChatGPT
````
